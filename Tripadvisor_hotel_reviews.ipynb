```python
selected_words = ['awesome', 'great', 'fantastic', 'amazing', 'love', 'horrible', 'bad', 'terrible', 'awful', 'wow', 'hate']
```

```python
def select_word(dic):
    dic=dict(dic)
    result=dict()
    for k in selected_words:
        if k in dic:
            result[k]=dic[k]
        else:
            result[k]=0
    return result
            
```

```python
products['selected_word_count']=products['word_count'].apply(select_word)
```

```python
products['selected_word_count']
```

```python
train_data,test_data=products.random_split(0.8,seed=0)
```

```python
seleted_sentiment_model=gl.logistic_classifier.create(train_data,features=['selected_word_count'],target='sentiment',validation_set=test_data)
```

```python
seleted_sentiment_model['coefficients'].print_rows(20)
```

```python
seleted_sentiment_model.evaluate(test_data)
# sentiment_model.evaluate(test_data)
```

```python
seleted_sentiment_model.show(view='Evaluation')
```
